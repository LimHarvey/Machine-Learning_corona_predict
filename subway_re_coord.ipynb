```python
# kakao API 사용해 주소 주소로 좌표 가져오기 test
# lat = 위도, lng = 경도
location = '서울 강북구 도봉로 지하 338'
url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(location)
headers = {"Authorization": "KakaoAK c148f7b7634196a3628b3b4cb2653943"}
result = json.loads(str(requests.get(url,headers=headers).text))
match_first = result['documents'][0]['address']
lat = float(match_first['y'])
lng = float(match_first['x'])

print(location,'\t', '위도:', lat, '경도:', lng)
```

```python
# kakao API 사용해 서울시 지하철역 좌표 추출
lat_list = []
lng_list = []
for i, addr in enumerate(list(df['역주소'])):
    location = addr
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(location)
    headers = {"Authorization": "KakaoAK c148f7b7634196a3628b3b4cb2653943"}
    result = json.loads(str(requests.get(url,headers=headers).text))
    match_first = result['documents'][0]['address']
    lat = float(match_first['y'])
    lng = float(match_first['x'])
    
    print(i+2, location,'\t', '위도:', lat, '경도:', lng)
    
    lat_list.append(lat)
    lng_list.append(lng)
```

```python
df['lat'] = lat_list
df['lng'] = lng_list
```
